```python
import numpy as np
import matplotlib.pyplot as plt

# Define lattice constant
a = 1.42  # Nearest-neighbor bond length in Angstroms
a0 = np.sqrt(3) * a  # Lattice constant

# Define real-space lattice vectors
a1 = np.array([a0, 0])
a2 = np.array([a0/2, np.sqrt(3)*a0/2])

# Compute reciprocal lattice vectors
b1 = (2 * np.pi / (3 * a)) * np.array([1, np.sqrt(3)])
b2 = (2 * np.pi / (3 * a)) * np.array([1, -np.sqrt(3)])

# Generate reciprocal lattice points
num_cells = 5  # Number of unit cells in reciprocal space
reciprocal_lattice = []

for i in range(-num_cells, num_cells + 1):
    for j in range(-num_cells, num_cells + 1):
        G = i * b1 + j * b2
        reciprocal_lattice.append(G)

reciprocal_lattice = np.array(reciprocal_lattice)

# Plot reciprocal lattice
plt.figure(figsize=(12, 6))
plt.scatter(reciprocal_lattice[:, 0], reciprocal_lattice[:, 1], c='red', s=50)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.xlabel("Reciprocal x (1/Å)")
plt.ylabel("Reciprocal y (1/Å)")
plt.title("Reciprocal Lattice of Graphene")
plt.gca().set_aspect('equal')
# plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```
