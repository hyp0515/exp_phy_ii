```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from scipy.optimize import curve_fit, minimize
from scipy.interpolate import make_interp_spline
```

```python
def sine_func(t, vpp, f, phi, vdc):
    return vpp * np.sin(2*np.pi*f*t*1e-3+phi) + vdc
```

## 0011

```python
df = pd.read_csv('./data/20250225-0011_02.csv', skiprows=4, header=None)
df.columns = ['Time', 'Channel A', 'Channel B']
```

```python
df['Time'] = pd.to_numeric(df['Time'], errors='coerce')
df['Channel A'] = pd.to_numeric(df['Channel A'], errors='coerce')
df['Channel B'] = pd.to_numeric(df['Channel B'], errors='coerce')
```

```python

plt.figure(figsize=(12, 9))
plt.plot(df['Time'], df['Channel A']*1e-3, label='Channel A')
# plt.plot(df['Time'], df['Channel B']*1e-3, label='Channel B')

plt.xlim((0, 100))
plt.ylim((-2, 2))

plt.xlabel('Time (ms)', fontsize=16)
plt.ylabel('Voltage (V)', fontsize=16)
plt.title('Experiment 2-1: No signal in both Channel A and Channel B', fontsize=16)

plt.legend()
plt.savefig('exp2-1.pdf', transparent=True)
```

```python
plt.close("all")
```

## 0010

```python
df = pd.read_csv('./data/20250225-0010_02.csv', skiprows=4, header=None)
df.columns = ['Time', 'Channel A', 'Channel B']
```

```python
df['Time'] = pd.to_numeric(df['Time'])
df['Channel A'] = pd.to_numeric(df['Channel A'])
```

```python
params, covariance = curve_fit(
    sine_func, 
    df['Time'], 
    df['Channel A'], 
    p0=[2, 2e3, 0, 1],
)
print(params)
```

```python
plt.figure(figsize=(12, 9))
plt.plot(df['Time'], df['Channel A'], label='Channel A')
plt.plot(df['Time'], sine_func(df['Time'], *params), label=f'Fitted curve: Vpp={params[0]:.2f}V, f={params[1]*1e-3:.2f}kHz, VDC={params[3]:.2f}V')
plt.xlim((0, 1.75))
plt.ylim((-2, 2))

plt.xlabel('Time (ms)', fontsize=16)
plt.ylabel('Voltage (V)', fontsize=16)
plt.title('Experiment 2-5: f=2kHz, Vpp=1V, VDC=1V', fontsize=16)

plt.legend(fontsize=14)
plt.savefig('exp2-5.pdf', transparent=True)
```

```python
plt.close("all")
```
